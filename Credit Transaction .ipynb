```python
# iterable function
def get_last_name(name):
    return name.split()[-1]
```

```python
""" 
6. New column (Name: ’CO’) to store the last word of company name. 
"""

last_name = balancesheet['Company Name'].map(get_last_name)
balancesheet['CO'] = last_name
balancesheet['CO']
```

```python
"""
7. Merge (inner) Ratings and BalanceSheet based on ’datadate’ and ’Global Com- pany Key’, 
and name merged dataset ’Matched’.
"""
Matched = pd.merge(ratings, balancesheet, on=['Data Date','Global Company Key'], how='inner')
```

```python
Matched.head()
```

```python
#8. Mapping
def numerical_rating(rate):
    if rate == 'AAA':
        return 0
    elif rate == 'AA+':
        return 1     
    elif rate == 'AA':
        return 2
    elif rate == 'AA-':
        return 3
    elif rate == 'A+':
        return 4
    elif rate == 'A':
        return 5
    elif rate == 'A-':
        return 6
    elif rate == 'BBB+':
        return 7
    elif rate == 'BBB':
        return 8
    elif rate == 'BBB-':
        return 9
    elif rate == 'BB+':
        return 10
    elif rate == 'BB':
        return 11
    else:
        return 12
    
Matched['Rate'] = Matched['S&P Domestic Long Term Issuer Credit Rating'].map(numerical_rating)
Matched['Rate']
```

```python
# 9. the rating frequency of company whose name end with ’CO’
Matched[Matched['CO']=='CO']['Rate'].value_counts()
```
